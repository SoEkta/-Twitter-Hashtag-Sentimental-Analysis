```python
corpus=[]
for i in range(len(x)):
    #print(x[i])
    review=re.sub('@\w* ',"", x[i])
    review=re.sub(r'(https|http)?:\/\/(\w|\.|\/|\?|\=|\&|\%)*\b', '', review)
    review=review.lower()
    corpus.append(review)
    
```

```python
corpus
```

```python
from afinn import Afinn     #affin is a sentiment library, gives score to each words in tweet
```

```python
af = Afinn()
```

```python
y=[]
for i in range(len(corpus)):
    a=af.score(corpus[i])
    y.append(a)
```

```python
y
```

```python
#printing number of total, positive, negative, and neutral tweet
n=0;
p=0;
ne=0;
for i in range(len(y)):
    if y[i]<0:
        n=n+1;
    elif y[i]>0:
        p=p+1;
    else:
        ne=ne+1;
t=len(y)
print('Total number of tweets extracted = ',t)
print('Total number of positive tweets = ',p)
print('Total number of neutral tweets = ',ne)
print('Total number of negative tweets = ',n)
```
